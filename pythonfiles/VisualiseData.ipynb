```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

alpha =0.05
df = pd.DataFrame.from_csv('formated\PumpMeasuresFormated.csv',sep=';')
# Add in the dataframe a new column representing the filterd current
df.ix[0, 'CurrentFilt']=df.ix[0, 'PumpCurrent']
for i in range(1, len(df)):
    df.ix[i, 'CurrentFilt'] = df.ix[i, 'PumpCurrent']*alpha + df.ix[i-1, 'CurrentFilt'] * (1-alpha)
```

```python
subset_a = df[df.Cycle==1].dropna()
subset_b = df[df.Cycle==2].dropna()
subset_c = df[df.Cycle==3].dropna()
subset_d = df[df.Cycle==4].dropna()
subset_e = df[df.Cycle==5].dropna()
subset_f = df[df.Cycle==6].dropna()
subset_g = df[df.Cycle==7].dropna()
subset_h = df[df.Cycle==8].dropna()
subset_i = df[df.Cycle==9].dropna()
subset_j = df[df.Cycle==10].dropna()
subset_k = df[df.Cycle==11].dropna()
```

Show the variation of the current of the pump  from the flow without filtering

```python
plt.figure(figsize=(10,10))
plt.scatter(subset_a.Flow, subset_a.PumpCurrent, s=120, c='b', label='Cycle1')
plt.scatter(subset_b.Flow, subset_b.PumpCurrent, s=60, c='r', label='Cycle2')
plt.scatter(subset_c.Flow, subset_c.PumpCurrent, s=60, c='m', label='Cycle3')
plt.scatter(subset_d.Flow, subset_d.PumpCurrent, s=60, c='y', label='Cycle4')
plt.scatter(subset_e.Flow, subset_e.PumpCurrent, s=60, c='c', label='Cycle5')
plt.scatter(subset_f.Flow, subset_f.PumpCurrent, s=60, c='k', label='Cycle6')
plt.legend()
```

Show the variation of the current of the pump  in function of the load of the entry (valve variation)

```python
plt.figure(figsize=(10,10))
plt.scatter(subset_a.EntryLoad, subset_a.PumpCurrent, s=120, c='b', label='Cycle1')
plt.scatter(subset_b.EntryLoad, subset_b.PumpCurrent, s=60, c='r', label='Cycle2')
plt.scatter(subset_c.EntryLoad, subset_c.PumpCurrent, s=60, c='m', label='Cycle3')
plt.scatter(subset_d.EntryLoad, subset_d.PumpCurrent, s=60, c='y', label='Cycle4')
plt.scatter(subset_e.EntryLoad, subset_e.PumpCurrent, s=60, c='c', label='Cycle5')
plt.scatter(subset_f.EntryLoad, subset_f.PumpCurrent, s=60, c='k', label='Cycle6')
plt.legend()
```

Show the current of the pump in function of the flow with filtering

```python
plt.figure(figsize=(10,10))
plt.scatter(subset_a.Flow, subset_a.CurrentFilt, s=120, c='b', label='Cycle1')
plt.scatter(subset_b.Flow, subset_b.CurrentFilt, s=60, c='r', label='Cycle2')
plt.scatter(subset_c.Flow, subset_c.CurrentFilt, s=60, c='m', label='Cycle3')
plt.scatter(subset_d.Flow, subset_d.CurrentFilt, s=60, c='y', label='Cycle4')
plt.scatter(subset_e.Flow, subset_e.CurrentFilt, s=60, c='c', label='Cycle5')
plt.scatter(subset_f.Flow, subset_f.CurrentFilt, s=60, c='k', label='Cycle6')
plt.legend()
```

Show the variation of the current of the pump in relation with the load of the entry (valve variation) with filtering

```python
plt.figure(figsize=(10,10))
plt.scatter(subset_a.EntryLoad, subset_a.CurrentFilt, s=120, c='b', label='Cycle1')
plt.scatter(subset_b.EntryLoad, subset_b.CurrentFilt, s=60, c='r', label='Cycle2')
plt.scatter(subset_c.EntryLoad, subset_c.CurrentFilt, s=60, c='m', label='Cycle3')
plt.scatter(subset_d.EntryLoad, subset_d.CurrentFilt, s=60, c='y', label='Cycle4')
plt.scatter(subset_e.EntryLoad, subset_e.CurrentFilt, s=60, c='c', label='Cycle5')
plt.scatter(subset_f.EntryLoad, subset_f.CurrentFilt, s=60, c='k', label='Cycle6')
plt.legend()
```
